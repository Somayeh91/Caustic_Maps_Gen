```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = open('GD1_ids.dat', 'r')
x = f.readlines()
```

```python
x = [(x[i].strip()).split('\t') for i in range(1,len(x))]
```

```python
x = np.asarray(x, dtype = float)
```

```python
df = pd.DataFrame({'ID': x[:,0],
                   'k': x[:,1],
                   'g': x[:,2],
                   's': x[:,3]})
len(df)
```

```python
df['macro_m'] = 1. / ((1-df.k)**2 - (df.g**2))
```

```python
plt.hist(df.macro_m.values)
plt.yscale('log')
```

```python
lim = 0.3
dff = df[(df.k < df.g+lim) & (df.k > df.g - lim)]
print(len(dff))
```

```python
dfff = df[(df.g < df.k+lim) & (df.g > df.k - lim)]
print(len(dfff))
```

```python
import random
id_random = random.sample(list(df.ID.values), 5000)
```

```python
df_random = df[df.ID.isin(id_random)]
```

```python
plt.figure(figsize = (15,15))
# plt.plot(df.k, df.g, '.', alpha = 0.1, markersize = 15)
plt.plot(df_random.k, df_random.g, '.', alpha = 0.1, markersize = 15)
```
